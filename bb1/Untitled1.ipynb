```python
import numpy as np 
import matplotlib.pyplot as plt
```

## 第一題

```python
v0 = np.array([ 3.0, 0])
origin = np.array([ 0.0, 0.0])
a = np.array([ -1.0 , -0.5])

# x極大值發生於 vx = 0 時
# 根據公式 v = v0 + at
#         x = origin + v0*t + 1/2at^2
t = np.abs(v0[0]/a[0])
v = v0 + a*t
x = origin + v0*t + 1/2*a*(t**2)

print ("位置在{}, 速度為{}, 時間為{}".format(x, v, t))
```

```python
# 初始值設定
path = []
origin = np.array([ 0, 0])
velocity = [] 

# 路徑
for T in np.arange(0, 5, 0.4):
    # 物理公式的部分
    p = origin + v0*T + 1/2*a*(T**2)
    v = v0 + a*T
    
    # 紀錄路徑
    velocity.append(v)
    path.append(p)
```

```python
# 轉換物件類別
path = np.array(path)
velocity = np.array(velocity)

# normalization, 方便觀看箭頭
velocity = velocity / np.linalg.norm(velocity)

# 作圖
plt.figure()
plt.scatter(path[:, 0], path[:, 1])
for i in range(path.shape[0]):
    plt.arrow(x=path[i, 0], y=path[i, 1], dx=velocity[i,0], dy=velocity[i,1], head_width=0.05, head_length=0.1, color='g')
    plt.arrow(x=path[i, 0], y=path[i, 1], dx=a[0], dy=a[1], head_width=0.05, head_length=0.1, color='r')
    plt.annotate(xy=path[i], text="{}".format(path[i]))
```

## 第二題

```python
# 初始值設定
_v = 10
g = np.array([0, -9.8])
paths = []
th = []

# 路徑
for theta in range(15, 90, 15): # 改變不同角度
    # 初始值
    path = []
    th.append(theta)
    v0 = np.array([_v*np.cos(theta*2*np.pi/360), _v*np.sin(theta*2*np.pi/360)])
    origin = np.array([ 0, 0])
    
    # 物理公式的部分
    for t in np.arange(0, 3, 0.01):
        p = origin + v0*t + 1/2*g*(t**2)
        
        # 一旦物體落地則停止紀錄路徑
        if (p[1]<0)&(p[0]!=0):
            break
        
        path.append(p)
        
    path = np.array(path)
    paths.append(path)
    
paths = np.array(paths)
th = np.array(th)
```

```python
# 作圖
plt.figure()
for i in range(paths.shape[0]):
    plt.plot(paths[i][:, 0], paths[i][:,1])
    plt.annotate(xy = paths[i][ 40, :], text = th[i])
```
